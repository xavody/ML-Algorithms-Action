# Gradient Descent Simulation

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plot_x = np.linspace(-1, 6, 141)
plot_y = (plot_x - 2.5) ** 2 - 1
```

```python
plt.plot(plot_x, plot_y)
```

```python
def dJ(theta):
    return 2 * (theta - 2.5)
```

```python
def J(theta):
    return (theta - 2.5) ** 2 - 1
```

```python
eta = 0.1
epsilon = 1e-8

theta = 0.0
while True:
    gradient = dJ(theta)
    last_theta = theta
    
    theta = theta - eta * gradient
    if (abs(J(theta) - J(last_theta)) < epsilon):
        break

print(theta)
print(J(theta))
```

```python
def gradient_descent_plot(initial_theta, eta, epsilon=1e-8):
    theta = initial_theta
    theta_history.append(theta)

    while True:
        gradient = dJ(theta)
        last_theta = theta

        theta = theta - eta * gradient
        theta_history.append(theta)
        
        if (abs(J(theta) - J(last_theta)) < epsilon):
            break

def plot_theta_history():
    plt.plot(plot_x, J(plot_x))
    plt.plot(np.array(theta_history), J(np.array(theta_history)), color='r', marker='+')
```

```python
eta = 0.1
theta_history = []
gradient_descent_plot(0., eta)
plot_theta_history()
```

```python
def J(theta):
    try:
        return (theta - 2.5) ** 2 - 1
    except:
        return float('inf')
```

```python
def gradient_descent_plot(initial_theta, eta, n_iters=1e4, epsilon=1e-8):
    theta = initial_theta
    theta_history.append(theta)
    i_ters = 0

    while i_ters < n_iters:
        gradient = dJ(theta)
        last_theta = theta

        theta = theta - eta * gradient
        theta_history.append(theta)
        
        if (abs(J(theta) - J(last_theta)) < epsilon):
            break
        
        i_ters += 1
```

```python
def gradient_descent(initial_theta, eta, n_iters=1e4, epsilon=1e-8):
    theta = initial_theta
    i_ters = 0

    while i_ters < n_iters:
        gradient = dJ(theta)
        last_theta = theta

        theta = theta - eta * gradient
        if (abs(J(theta) - J(last_theta)) < epsilon):
            break
        
        i_ters += 1
    
    return theta
```
